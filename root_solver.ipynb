# Problem #1 Solving for Roots

The equation to be solved is $f(z)= z^5 -1 $

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
import numpy as np
```

##a) How many distinct roots?

There is only one real root, at $z=1$ which can be seen in the plot below. The other 4 roots must be in the complex plane. And given that complex roots occur in conjugate pairs, there must be atleast 2 distinct roots in the complex plane. To find whether the other two are distinct, we use the idea of discriminant.

For the given 5th degree polynomial, $\Delta=3125 \neq 0$. Therefore, there are no multiple roots in the complex plane. This must mean there are two more distinct complex roots

To answer the original question, how many distinct roots does the equation have- **the answer is 5**

```python
z=np.linspace(-1.4,1.4,100,endpoint=True)
y= (z**5) - 1

fig, ax = plt.subplots(1, 1)
ax.get_xaxis().set_minor_locator(ticker.AutoMinorLocator())
ax.get_yaxis().set_minor_locator(ticker.AutoMinorLocator())
ax.grid(b=True, which='major', color='w', linewidth=1.0)
ax.grid(b=True, which='minor', color='w', linewidth=0.5)
plt.plot(z,y)
plt.show()
```

##b) Code for Newton Raphson solver

###Non vectorized test solver

```python
def f(x):
    return x**5 - 1

def fdash(x):
    return 5*x**4

def solver(x0):
    fs=[]
    xs=[]
    fs.append(f(x0))
    xs.append(x0)
    tol=1000.0
    n_iter=0
    while(tol> .00001 and n_iter<30000):
        try:
            xnew= x0 - f(x0)/fdash(x0)
        except:
            fs.append(np.NaN)
            xs.append(np.NaN)
            n_iter+=1
            return xs,fs,n_iter
        tol= abs(xnew-x0)
        fs.append(f(xnew))
        xs.append(xnew)
        x0=xnew
        n_iter+=1
    return xs,fs,n_iter
```

###Vectorized implementation

```python
def solver_vec(x0): #x0 is a matrix
    tol=np.float64(1000) #using static typing
    n_iter=np.int64(0)
    tol_limit= np.float64(0.00001)
    n_limit= np.int64(30000)
    while(tol> tol_limit and n_iter<n_limit):
        xnew= x0 - f(x0)/fdash(x0)
        tol= np.amax(np.abs(xnew-x0))
        x0=xnew
        n_iter+=1
    return xnew, n_iter
```

####Testing performance of both

```python
%%timeit -r 20
for el in np.linspace(-1j,1j,50):
    a,b,c=solver(el)
```

```python
%timeit -r 20 x,n_iter=solver_vec(np.linspace(-1j,1j,50))
```

The vectorized version performs **44%** faster than the non-vectorized version.

###Testing the solver on the real plane

```python
zs,fs,n_iter=solver(x0=2.0)
zs=np.array(zs)
fs=np.array(fs)
print "final val of f(z)=",fs[-1]," at z=", zs[-1], "in n iter=", n_iter
print "The first (and only real) root is ", zs[-1]
```

###Finding the remaining roots

```python
x_inits= np.linspace(-1j,1j,num=50,endpoint=True)
roots=set()

for x_init in x_inits:
    zs,fs,n_iter=solver(x_init)
    finalz= zs[-1]
    finalz= round(finalz.real,3)+ round(finalz.imag,3)*1.0j
    roots.add(finalz)
print "The roots of the equation are:"
for el in roots:
    print el
```

##3) Fractal pattern

```python
print "Number of divisions needed between two sides of the square=", (1- (-1))*1024
r= np.linspace(-1,1,2048,endpoint=True)
s= np.linspace(-1j,1j,2048,endpoint=True)
rr,ss= np.meshgrid(r,s)
x_inits= rr+ss
root_2d= np.zeros_like(x_inits)
print "We will be evaluating the solver over a region of size", x_inits.shape
```

###Timing both non vectorized and vectorized versions of the code

```python
%%time

for index,val in np.ndenumerate(x_inits):
    root_2d[index]= solver(val)[0][-1] #indexing for converged z value
```

```python
%%time
root_2d,n_iter=solver_vec(x_inits)
```

###Plotting the roots
The grid we just created are initial points for the Newton-Raphson solver. We've just found the root at each of these points. Let's see how it looks

```python
root_2d_rnd= np.round(root_2d,decimals=3)
roots=np.unique(root_2d_rnd)

cols=np.arange(50,255,5)
for i,root in enumerate(roots):
    root_2d_rnd[root_2d_rnd==root]= cols[i]
col_arr= root_2d_rnd.real

plt.imshow(col_arr,cmap="YlGnBu",interpolation="bicubic",)
frame1=plt.gca()
frame1.axes.get_xaxis().set_visible(False)
frame1.axes.get_yaxis().set_visible(False)
plt.title("Fractal!!")
plt.show()
```

Each homogeneous patch of color in the above plot represents starting points that all led the solver to converge to the same root. It seems that certain bands of initial points bias the Newton-Raphson to converge at the same locations. We can also see repeating, recursive patterns- fractals!
