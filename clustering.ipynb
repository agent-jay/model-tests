##Modeling

```python
from scipy.spatial.distance import cdist
from sklearn.metrics import silhouette_score

#modl
def fitK(X, K):
    kmn= KMeans(n_clusters=K, n_init=25,max_iter=500)
    ylab=kmn.fit_predict(X)
    SSE=0  #Sum of squared errors (or) Variance
    for k in range(K):
        Ck= X[ylab==k]
        cent_k=np.reshape(kmn.cluster_centers_[k],(1,X.shape[1]))
        dist= cdist(Ck,cent_k,"euclidean")
        withinCk_SSE= np.sum(dist**2)
        SSE+=withinCk_SSE
    
    return SSE, silhouette_score(X, ylab, metric='euclidean')

#diagnostic plots

def elbow_plot(SSE, K):
    plt.plot(range(2,K+1),SSE.tolist()[0], "o-")
    plt.xlim(0,K+2)
    plt.ylim(0, SSE.max()*1.1)
    plt.xticks(range(0,K+2))
    plt.grid(True)
    plt.xlabel('Number of clusters')
    plt.ylabel('Sum of Squared Errors')
    plt.title('Elbow for KMeans clustering')
    plt.show()
    
def sil_plot(sil_score,K):
    plt.plot(range(2,K+1),sil_score.tolist()[0], "o-")
    plt.xlim(0,K+2)
    plt.ylim( 0, sil_score.max()*1.1)
    plt.xticks(range(0,K+2))
    plt.grid(True)
    plt.xlabel('Number of clusters')
    plt.ylabel('Silhouette score')
    plt.title('Average Silhouette score for different K values')
    plt.show()
```

###1.Original data (only id var removed)

The clustering is done on the dataset with the ID variable removed. It is possible that the algorithm starts finding meaningless patterns using the ID variable value. It is also possible however that it uncovers some latent structure of the data- perhaps the data is sorted in some order. It is worth redoing the below exercise without removing the ID variable

```python
K=20
Xmod=cust.ix[:,1:]
Xmod= scale(Xmod)
print Xmod.shape

SSE=np.zeros((1,K-1))
sil_score=np.zeros((1,K-1))
```

```python
%%time

for i in range(2,K+1):
    SSE[0,i-2], sil_score[0,i-2]= fitK(Xmod,i)
    print "for k= ",i,"SSE= ",SSE[0,i-2],"Silhouette score= ",sil_score[0,i-2]
```

```python
elbow_plot(SSE,K)
sil_plot(sil_score,K)
```

The elbow plot is used to show the variance reduction with the addition of each new cluster. In the ideal case, we should see a distinct *elbow* in the graph, where the reduction in variance goes from dramatic, to incremental. That isn't the case with the above plot, however we can say that the elbow should exist between $K=8$ and $K=12$ as this is the region in which the we see the transition in variance from high to low

In order to find the ideal K and break the tie, we use another metric called the **Silhouette Score**. It is defined for a sample as: $$b-a/max(a,b)$$ where, **a=** mean intra-cluster distance 
**b=**distance between a sample and the nearest cluster

It is a score $\in(-1,1)$, with higher scores indicating greater distance from other clusters, and smaller intra-cluster distance.

From the above plot, it can be seen that $K=11$ is the right choice- it minimizes K while increasing the Silhouette score



###2. Modified data- ID AND boolean variables removed

When the data was inspected, it was noticed that there are 3 variables- MEMBER_CLUB, POINTS_COLLECTOR and REGISTRATION_B that contain 
boolean values. The Euclidean distance (used in Kmeans) between True and False values has no meaning and can skew results. Hence, we
redo the above excercise removing these variables. Also note that the features are scaled as before

```python
K=20
Xmod= cust[[i for i in range(18) if i not in [0,11,12,17]]]
Xmod= scale(Xmod)
print Xmod.shape
SSE_white=np.zeros((1,K-1))
sil_score_white=np.zeros((1,K-1))
```

```python
%%time
for i in range(2,K+1):
    SSE_white[0,i-2], sil_score_white[0,i-2]= fitK(Xmod,i)
    print "for k= ",i,"SSE= ",SSE_white[0,i-2],"Silhouette score= ",sil_score_white[0,i-2]
```

```python
elbow_plot(SSE_white, K)
sil_plot(sil_score_white, K)
```

While the elbow plot is similar to that of the previous version, we notice that right off the bat, Silhouette scores have jumped indicating that our choice to drop the boolean variables was well informed. Also, it is apparent that for this transformed data, $K=12$ is the right choice.

While performing this exercise, each iteration would cause the results to oscillate (due to the stochastic nature of the process) with **K** swinging between 10 and 12. Unless a very clear difference is present in the different classes, this behaviour is acceptable from Kmeans. The takeaway is that the ideal **K** is in a range of 10-12, and domain information could help break the tie.

##Visualizing the clusters using PCA

The given data has 14 dimensions and the clusters cannot be visualized. Enter PCA. Using this dimension reduction technique, we can reduce our 14-dimensional data to a 2/3- dimensional one and see if the clusters actually do separate the data out.

First we check the distribution of the data across clusters (K=12)

```python
from collections import Counter
kmn= KMeans(n_clusters=12, n_init=20, max_iter=500)
ylab=kmn.fit_predict(Xmod)

a=Counter(ylab)
#print a.viewitems()

plt.title("Distribution across clusters")
plt.hist(ylab,bins=12)
plt.show()
```

```python
from sklearn.decomposition import PCA
pca= PCA(2)

Xproj= pca.fit_transform(Xmod)
plt.scatter(Xproj[:, 0], Xproj[:, 1], c=ylab, edgecolor='none', alpha=0.5,cmap='jet')
plt.colorbar();
```

```python
pca= PCA(3)
Xproj= pca.fit_transform(Xmod)

fig = plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.scatter(Xproj[:,0], Xproj[:,1], Xproj[:,1], c=ylab, cmap='jet', alpha=.6)
plt.show()
```

We can see that there is some pattern in the clusters and KMeans has separated them out to some extent. But we also note the dense nature of the data, earlier indicated by the low ($<0.5$) Silhouette scores. The nature of the data is that a clustering algorithm will have trouble separating out the different classes, without transforming it into a more meaningful space or use a better clustering technique
