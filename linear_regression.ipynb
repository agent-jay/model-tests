## 1. Modeling using Gradient Descent

```python
def linreg(X,y, alpha=.5 ):
    theta= np.ones((X.shape[1],1))
    y= dat.A.reshape(X.shape[0],1)
    pass

def hyp(X,theta):
    return np.dot(X,theta)

def cost(X, theta, y):
    return np.sum((hyp(X,theta)-y)**2) / (2.0*X.shape[0])

def single_iter(X,theta,y, alpha):
    return theta - (alpha/1.0*X.shape[0])* np.dot( X.T, hyp(X,theta)-y )

def grad_desc( X, theta, y, alpha=.00001):
    iter_n=0
    J_l=[]
    
    #1st iter
    oldJ=None
    J= cost(X,theta,y)
    theta= single_iter(X,theta,y, alpha)
    J_l.append(J)
    
    n_iter=1
    
    while(oldJ==None or (oldJ-J) > .0000001 and n_iter < 30000):  #setting convergence criteria
        oldJ=J
        J= cost(X,theta,y)
        theta= single_iter(X,theta,y, alpha)
        J_l.append(J)
        n_iter +=1
            
    return theta, J_l
```

###Running 100 iterations of gradient descent, each time starting with a different theta

```python
conv_val=[]

for i in range(100):
    theta= np.random.rand(X.shape[1],1)
    theta, Js= grad_desc(X,theta, y, alpha=.00001)
    if conv_val:
        if Js[-1]< conv_val[1][-1]:
            conv_val=[theta, Js]
    else:
        conv_val=[theta, Js]

plt.plot(conv_val[1], "-")
print "Final val of cost function", conv_val[1][-1]
plt.show()
```

##3. Is this a good model?

While we have built a decent model, it is fairly basic. The following can be done to improve it:
1. **Test-Train-split**- This model was built on the entire data. Ideally, the model should be built on training data and tested on holdout (test) data. This will ensure that the model isn't memorizing the dataset, and is capable of generalizing to out of sample data
2. **Regularization**- Adding a regularization term to the cost function is an additional step we can take in the model building phase to avoid overfitting. Regularization penalizes features that have large values of $\beta$
3. **Scaling**- The inputs are of different scales. By scaling the features appropriately, convergence during gradient descent can be sped up (not for Normal equation)
4. **Correlation**- As can be seen in the correlation and pairplots below, the variables C and D are highly correlated. One of these can be removed and the model be rebuilt.
5. **Polynomial features**- It is possible that transforming the feature space (quadratic,...) could create better features.
